```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def compute_E(spins):
    roll_x = np.roll(spins, 1, axis = 1)
    roll_y = np.roll(spins, 1, axis = 0)
    return -1*((spins * roll_x).sum() + (spins * roll_y).sum())
```

```python
my_spins = np.ones((8,8))

def monte_carlo_move(spins, kbT, steps):
    avg_spins = []
    plot = []
    plot.append(np.sum(spins)/np.size(spins))

    E_before = compute_E(spins)
    
    for i in range(0,steps):
        pick_spin_x = np.random.randint(spins.shape[0])
        pick_spin_y = np.random.randint(spins.shape[1])
        spins[pick_spin_x, pick_spin_y] *= -1
        E_after = compute_E(spins)
            
        if E_after < E_before:
            E_before = E_after
        else:
            rand = np.random.random()
            boltz = np.exp(-(E_after-E_before)/kbT)
            if rand < boltz:
                E_before = E_after
            else:
                spins[pick_spin_x, pick_spin_y] *= -1
            plot.append(np.sum(spins)/np.size(spins))

    #heat map
    plt.imshow(spins)
    plt.show()
    
    #graph steps
    fig, ax = plt.subplots()
    ax.plot(plot)
    ax.set_ylabel('Average Magnetization')
    ax.set_xlabel('Monte Carlo steps')
    plt.show()
    return spins, E_after
    
```

```python
#Question c and d.
spins1, E_after1 = monte_carlo_move(my_spins, 1,100000)
spins2, E_after2 = monte_carlo_move(my_spins, 2,100000)    
spins3, E_after3 = monte_carlo_move(my_spins, 4,100000)
```

```python

def checkerboard(shape):
    return 2*(np.indices(shape).sum(axis=0) % 2) - 1
check_spin = checkerboard((100, 100))

spins1, E_after1 = monte_carlo_move(check_spin, 1,100000)
spins2, E_after2 = monte_carlo_move(check_spin, 2,100000)    
spins3, E_after3 = monte_carlo_move(check_spin, 4,100000)
```
