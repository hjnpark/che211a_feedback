```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
kb = 1.3806452e-23 #unit in J/k
Na = 6.02214e23 #unit in 1/mol
mass = (84 / 1000)/Na #unit in kg
T = 300 #unit in K 
```

```python
# speed distribution
def MB_speed(v):

    beta = 1/(kb*T)
    term_1 = 4 * np.pi * v ** 2 
    term_2 = mass / (2*np.pi*kb*T)
    term_3 = -beta * mass * v ** 2 / 2
    Pv = term_1 * (term_2) ** (3/2) * np.exp(term_3)

    return Pv 

speed = np.arange(0,1000)
distProb = MB_speed(speed)
mean = np.sqrt((8 * kb * T)/(np.pi * mass))
SD = np.sqrt((kb*T/mass)*(3-8/np.pi))

print ( "The standard deviation of speed is", SD, "m/s")
print ( "The average speed is", mean, "m/s")

#plot
plt.plot(distProb)
plt.xlabel("speed(m/s)")
plt.ylabel("probability")
plt.show()
```
