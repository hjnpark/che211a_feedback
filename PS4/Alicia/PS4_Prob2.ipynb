```python
import matplotlib.pyplot as plt
import numpy as np
kb = 1.380649e-23 
Na = 6.022e23
```

```python
mass = (84 / 1000)/Na
temp = 300
def vel_dist(vel, temp):
    beta = 1/(kb*temp)
    term1 = 4 * np.pi * vel**2
    term2 = (mass/(2 * np.pi * kb * temp))**(3/2)
    term3 = np.exp(-(beta * mass * vel**2)/2)
    prob = term1 * term2 * term3
    
    return prob
```

```python
speed = np.arange(0, 1000)
mean = np.sqrt((8 * kb * temp)/(np.pi * mass))
SD = np.sqrt(((kb * temp)/mass) * (3 - (8/np.pi)))
```

```python
dist300 = vel_dist(speed, 300)
```

```python
plt.plot(dist300) 
plt.plot(mean, 0, marker = "o" , markersize=5)
plt.plot(mean + SD, 0, marker = "o" , markersize=5)
plt.plot(mean - SD, 0, marker = "o" , markersize=5)
plt.plot(mean + 2*SD, 0, marker = "o" , markersize=5)
plt.plot(mean - 2*SD, 0, marker = "o" , markersize=5)
plt.ylabel("Probability")
plt.xlabel("Velocity in m/s")
plt.show()
```

The orange dot representes the calculated mean, the red and green points are one standard deviation away from the mean
which represents around 68% of the probability distribution area. The brown and purple points are 2 standard deviations
away which represent about 95% of the area under the probability distribution.
