```python
def Energy(spins):
    rolly = np.roll(spins, 1, axis = 0)
    rollx = np.roll(spins, 1, axis = 1)
    return -1*((spins*rollx).sum() + (spins*rolly).sum())
```

```python
trial = np.ones((8,8))
```

```python
trialE = Energy(trial)
trialE
```

```python
def monte_carlo_move(spins, E_before):
    #E_before = Energy(spins)
    pick_spin_row = np.random.randint(spins.shape[0])
    pick_spin_column = np.random.randint(spins.shape[1])
    spins[pick_spin_row, pick_spin_column] *= -1
    
    rollrowpos = np.roll(spins, 1, axis = 0)
    rollrowneg = np.roll(spins, -1, axis = 0)
    rollcolumnpos = np.roll(spins, 1, axis = 1)
    rollcolumnneg = np.roll(spins, -1, axis = 1)
    E_after = E_before - 2*(spins[pick_spin_row, pick_spin_column] * rollrowpos[pick_spin_row, pick_spin_column] + spins[pick_spin_row, pick_spin_column] * rollcolumnpos[pick_spin_row, pick_spin_column] + spins[pick_spin_row, pick_spin_column] * rollrowneg[pick_spin_row, pick_spin_column] + spins[pick_spin_row, pick_spin_column] * rollcolumnneg[pick_spin_row, pick_spin_column])
    if E_after < E_before:
        return E_after
    else:
        rand = np.random.random()
        boltz = np.exp( -(E_after - E_before)/kbT)
        if rand < boltz:
            return E_after
        else:
            spins[pick_spin_row, pick_spin_column] *= -1
            return E_before
```

```python
kbT = 1
trialT1 = np.ones((8,8))
E_before = Energy(trialT1)
for i in range(100000):
    E_before = monte_carlo_move(trialT1, E_before)
    

plt.imshow(trialT1)
plt.colorbar()
plt.show()
```

```python
kbT = 2
trialT2 = np.ones((8,8))
E_before = Energy(trialT2)
for i in range(100000):
    monte_carlo_move(trialT2, E_before)

plt.imshow(trialT2)
plt.colorbar()
plt.show()
```

```python
kbT = 4
trialT4 = np.ones((8,8))
E_before = Energy(trialT4)
for i in range(100000):
    monte_carlo_move(trialT4, E_before)

plt.imshow(trialT4)
plt.colorbar()
plt.show()
```

```python
def checkerboard(shape):
    return 2*(np.indices(shape).sum(axis=0) % 2) - 1
```

```python
kbT = 1
trialT1 = checkerboard((100,100))
E_before = Energy(trialT1)
for i in range(100000):
    monte_carlo_move(trialT1, E_before)

plt.imshow(trialT1)
plt.colorbar()
plt.show()
```

```python
kbT = 2
trialT2 = checkerboard((100,100))
E_before = Energy(trialT2)
for i in range(100000):
    monte_carlo_move(trialT2, E_before)

plt.imshow(trialT2)
plt.colorbar()
plt.show()
```

```python
kbT = 4
trialT4 = checkerboard((100,100))
E_before = Energy(trialT4)
for i in range(100000):
    monte_carlo_move(trialT4, E_before)

plt.imshow(trialT4)
plt.colorbar()
plt.show()
```

There are pockets that are pretty visible at kbT = 1. This is observed because the Energy is lowered
by increasing the number of same-spin interactions in a system with no magentic field. When one spin is flipped and is the same as its neighbors, the probability in that area for another to flip to allign with those around it also increases because it is lowering the energy.
