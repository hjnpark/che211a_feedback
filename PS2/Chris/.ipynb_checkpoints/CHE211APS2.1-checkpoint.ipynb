```python
import numpy as np
from matplotlib import pyplot as plt
from random import uniform as ru
import math
```

```python
kb = 1.38064852e-23
hw = 11.963
def HarmonicOscillatorEntropy(N,E):
    Term1 = (E/hw)*np.log(1+N*hw/E)
    Term2 = N*np.log(1+E/(N*hw))
    S = kb*(Term1+Term2)
    return S 
Energies1 = np.arange(1,5001)
Energies2 = np.arange(1,301)
Entropies1 = HarmonicOscillatorEntropy(1000,Energies1)
Entropies2 = HarmonicOscillatorEntropy(100,Energies2)
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3)
fig.set_size_inches(8,8)
ax1.plot(Energies1,Entropies1,label='Entropy')
ax3.plot(Energies2,Entropies2,label='Entropy2')
ax1.set_ylabel('Entropy (J/K)')
MaxLog = np.max(Entropies1)/kb/np.log(10)
MaxLogBase = 10**(MaxLog - int(MaxLog))

NumStates = np.exp((Entropies1 - np.max(Entropies1))/kb)
ax2.plot(Energies1,NumStates,label = '# States (Relative)')

ax2.set_ylabel('# States / %.1fe%i' % (MaxLogBase, int (MaxLog)))
ax2.set_xlabel('Energy (kj/mol)')

fig.savefig('entropy_ho1.pdf')
```

```python
def LJpot(rij, epsilon=1.77, sigma=4.10):
    sr = sigma/rij
    lj = 4*epsilon*(sr**12-sr**6)
    return lj

def derivLJpot(rij, epsilon=1.77, sigma=4.10):
    sr = sigma/rij
    dlj = epsilon*( -48. * sr**12/rij + 24*sr**6/rij )
    return dlj
distance_array= np.linspace(1.0,10.0,900)
F_array = -derivLJpot(distance_array)
EQ = distance_array[np.argmin(abs(F_array))]
potential_array = LJpot(distance_array)
NF = derivLJpot(6, epsilon=1.77, sigma=4.10)
```

```python
def calc_force_energy(xenon_pos):
    E_total = 0
    Force_1 = np.zeros(xenon_pos.shape)
    for i in range(len(xenon_pos)): 
        for j in range(len(xenon_pos)):
            if j>i:
                vector = xenon_pos[i]-xenon_pos[j]
                dis = np.linalg.norm(vector)
                f1 = derivLJpot(dis, epsilon=1.77, sigma=4.10)*-1*vector/dis
                Force_1[i] += f1
                Force_1[j] += -f1
                E_total+=LJpot(dis, epsilon=1.77, sigma=4.10)
    return Force_1, E_total

def optimize(n,stepsize, stepnumber):
    cube = []
    for i in range(n):
        for j in range(n):
            for k in range(n):
                cube.append([i,j,k])
    cube = np.array(cube)*EQ
    Energy_list = []
    x_list = []
    for i in range(stepnumber):
        Force_array, Energy = calc_force_energy(cube)
        Energy_list.append(Energy)
        x_list.append(i)
        cube += Force_array/(np.linalg.norm(Force_array)) * stepsize
    return cube, x_list,Energy_list

_, x_list, Energy_list = optimize(3, 0.01, 3000) # This is for stepsize of 0.01. You can try 0.1 and 0.001 now.

plt.plot(x_list, Energy_list)
```
