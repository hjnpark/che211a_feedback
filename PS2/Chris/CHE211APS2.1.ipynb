```python
import numpy as np
from matplotlib import pyplot as plt
from random import uniform as ru
import math
```

```python
kb = 1.38064852e-23
hw = 11.963
def HarmonicOscillatorEntropy(N,E):
    Term1 = (E/hw)*np.log(1+N*hw/E)
    Term2 = N*np.log(1+E/(N*hw))
    S = kb*(Term1+Term2)
    return S 
Energies1 = np.arange(1,5001)
Energies2 = np.arange(1,301)
Entropies1 = HarmonicOscillatorEntropy(1000,Energies1)
Entropies2 = HarmonicOscillatorEntropy(100,Energies2)
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3)
fig.set_size_inches(8,8)
ax1.plot(Energies1,Entropies1,label='Entropy')
ax3.plot(Energies2,Entropies2,label='Entropy2')
ax1.set_ylabel('Entropy (J/K)')
MaxLog = np.max(Entropies1)/kb/np.log(10)
MaxLogBase = 10**(MaxLog - int(MaxLog))

NumStates = np.exp((Entropies1 - np.max(Entropies1))/kb)
ax2.plot(Energies1,NumStates,label = '# States (Relative)')

ax2.set_ylabel('# States / %.1fe%i' % (MaxLogBase, int (MaxLog)))
ax2.set_xlabel('Energy (kj/mol)')

fig.savefig('entropy_ho1.pdf')
```
